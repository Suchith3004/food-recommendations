```python
import pandas as pd
import gzip
import json
```

```python
def parse(path):
    g = gzip.open(path, 'rb')
    for l in g:
        yield json.loads(l)

def getDf(path):
    i = 0
    df = {}
    for d in parse(path):
        df[i] = d
        i += 1
    return pd.DataFrame.from_dict(df, orient='index')
```

```python
df = getDf('./Grocery_and_Gourmet_Food_5.json.gz')
```

```python
df_cleaned = df.drop(columns=['reviewTime', 'unixReviewTime', 'style', 'image'])
```

```python
df_cleaned = df_cleaned[df_cleaned['verified'] == True]
```

```python
df_cleaned = df_cleaned.drop(columns='verified')
```

```python
to_csv = df_cleaned.to_csv('./gourmet_foods.csv', index=False)
```
